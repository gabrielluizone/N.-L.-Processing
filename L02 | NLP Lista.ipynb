> **Análise:** Não, ele não acertou todas. Teos algumas palavras que ele não souber identificar, como `BigTechs`, `deles` e `precisei`. Teve palavras com classificação incorreta como `do` e `pela`que foi classificado como NPROP que pode ser substantivo próprio

```python
sentenca = 'Em 15 anos trabalhando em Data Science e 10 deles \
            em BigTechs eu vi que a maioria ali está apenas \
            correndo atrás do próprio rabo e pela graça de Deus \
            eu sempre tive a liberdade de fazer o que eu queria \
            e nunca precisei seguir o mesmo caminho.'

sent_tokens = nltk.word_tokenize(sentenca)

# Sentenças de Treino
sentencas_treino = nltk.mac_morpho.tagged_sents()
etiquetador = nltk.UnigramTagger(sentencas_treino, backoff = nltk.DefaultTagger('N')) # Pra ter um nulo)

# Viewww
etiquetador.tag(sent_tokens)
```

7. **Retorne as classes gramaticais de todas as palavras do seu corpus**

```python
etiquetado = etiquetador.tag(tokens)
etiquetado
```
