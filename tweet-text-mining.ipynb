## Tweet Text Mining
> Using Tweets for text mining with nltk

---
- Curpus dos tweets
- Leitura `RAW` dos textos
- Seleção de palavras
- Remoção de stopwords
- Visualização & Frequência

```python
# Importação das bibliotecas
from nltk.corpus import PlaintextCorpusReader # Leitor de corpus especificos para o texto
from nltk.corpus import stopwords
import nltk

# Formatação & Markdown
from IPython.display import Markdown as md

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
import string

nltk.download('stopwords');
```

```python
# Criação de um corpus lendo os arquivos
corpus = PlaintextCorpusReader('/kaggle/input/text-mining-one', '.*', encoding = "ISO-8859-1")

# Leitura dos arquivos e percorrer os registros e seus nomes
files = corpus.fileids()
files[:7]
```

```python
# Acesso ao arquivo em especifico
text = corpus.raw(files[0])
display(text)

# Acesso a todas as palavras de todos os arquivos
alltext = corpus.raw()
allword = corpus.words()
```

```python
# Stopwords em Inglês
stops = stopwords.words('english')
#stops = stopwords.words('portuguese')
```

```python
# Definição das cores que serão utilizadas na nuvem de palavras
mapa_cores = ListedColormap(['#d45757', # Vermelho
                             '#5787d4', # Azul
                             '#57d480', # Verde
                             '#d4a057' # Laranja
])

# Criação da nuvem de palavras, com no máximo 100 palavras e utilizando as stop words
cloud = WordCloud(background_color = '#202124',
                  colormap = mapa_cores,
                  stopwords = stops,
                  max_words = 100)

# Criação e visualização da nuvem de palavras
cloud.generate(alltext)

plt.figure(figsize=(20,10), dpi=300)
plt.imshow(cloud);
```

```python
words_without_punctuation = [word for word in allword if (word not in stops) and (word not in string.punctuation)]
freq = nltk.FreqDist(words_without_punctuation)
freq
```

```python
freq.most_common(100)
```
